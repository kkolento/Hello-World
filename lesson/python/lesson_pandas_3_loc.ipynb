# Pandas数据索引

```python
import pandas as pd
path='J:/Learning/opencv/new/hw1/train.csv'
sample=pd.read_csv(path,encoding = 'big5')
sample.head(5)
```

## .loc索引

可以将列标签打印出来

```python
print(sample.columns)
```

```python
data1=sample.loc[sample['日期']=='2014/1/1','日期':'0']#逗号分隔行列,冒号加空格代表取相应方向所有数据
print(data1)
type(data1)
```

此时数据仍为dataframe类型，可以在loc语句后加**.values**，将数据类型直接转换为ndarray数组形式！便于后续各种计算

```python
data2=sample.loc[sample['日期']=='2014/1/1','日期':'0'].values
print(data2)
type(data2)
```

## .iloc索引

**1. 如果只使用一个参数，代表索取对应行的数据**

```python
sample.iloc[0].values
```

**2. 索引多行多列（离散）**

```python
sample.iloc[[0,3,4],[0,1,3]]
```

**3. 索引多行多列（连续）**

***左闭右开！！！*** 可以看到第4行和第4列并未被选中（起始行列位置为0）

```python
sample.iloc[1:4,0:4]
```

# .iloc与.loc索引对比

为了直观显示，我们对数据进行一下排序

pandas中 **.sort_index(ascending=False)** 可以按标签降序重新排列数据

```python
data3=sample.sort_index(ascending=False)
data3.head(3)
```

```python
data3.tail(3)
```

```python
dataloc=data3.loc[0,:].values
print(dataloc)
datailoc=data3.iloc[0].values
print(datailoc)
```

可以看到，iloc是按当前数据的实际所处行数索引，loc是按标签索引
